## Check PySpark in container

```python
import pyspark
print('PySpark version:', pyspark.__version__)
print('Spark location:', pyspark.__file__)
```

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .master('local[*]') \
    .appName('zoomcamp-test') \
    .getOrCreate()

print('Spark UI:', spark.sparkContext.uiWebUrl)
print('Spark version:', spark.version)
```

```python
data = range(1, 10001)
distributed = spark.sparkContext.parallelize(data)
result = distributed.filter(lambda x: x < 10).collect()
print('Result (should be 1..9):', result)
```

```python
# Test DataFrame
df = spark.createDataFrame([
    (1, 'Alice', 25),
    (2, 'Bob',   30),
    (3, 'Carol', 22),
], ['id', 'name', 'age'])

df.show()
df.printSchema()
```

```python
# Save to Parquet
df.write.mode('overwrite').parquet('/workspace/test_output')
print('Parquet written to /workspace/test_output')

# Read back
df2 = spark.read.parquet('/workspace/test_output')
df2.show()
```

```python
# Spark UI available at http://localhost:4040
print('Open in browser: http://localhost:4040')
spark.stop()
```
